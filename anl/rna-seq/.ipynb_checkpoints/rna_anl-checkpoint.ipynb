```python
import pandas as pd
import decoupler as dc  # we need decoupler 2.0 version at least
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import umap
import matplotlib.pyplot as plt
import seaborn as sns
import anndata as ad
```

```python
# Load batch corrected counts and metadata
mat = pd.read_csv("/mnt/c/Users/Roger/Downloads/adjusted_counts.tsv", sep="\t", index_col = 0)
metadata = pd.read_csv("/mnt/c/Users/Roger/Downloads/metadata.tsv", sep="\t")
```

```python
# filter metadata as we want
metadata = metadata[metadata['run'] == 72]
metadata = metadata[metadata['time'] == 'four']
metadata = metadata[metadata['treatment'] != 'TNFa']
```

```python
# remove rows that have all 0s
mat = mat.loc[~(mat == 0).all(axis=1)]
```

```python
df_T = mat.transpose()  # Now rows = samples and columns = genes
```

```python
four = df_T[df_T.index.isin(metadata['sampleID'])]
```

```python
#Create AnnData object
adata = ad.AnnData(X=four.values, obs=metadata, var=pd.DataFrame(index=four.columns))   #or df_T

adata.obs_names = four.index       # sample IDs
adata.var_names = four.columns     # gene symbols
```

```python
# see our experimental desing (that is very useful when you have more conditions)
dc.pl.obsbar(
    adata=adata,
    y='treatment',
    hue='time',
    figsize=(3, 3)
)
```

```python
# arbitrary and accepted thresholds.
dc.pl.filter_by_expr(
    adata=adata,
    group='treatment',
    min_count=10,
    min_total_count=15,
    large_n=10,
    min_prop=0.7,
)
```

```python
dc.pp.filter_by_expr(
    adata=adata,
    group='treatment',
    min_count=10,
    min_total_count=15,
    large_n=10,
    min_prop=0.7,
)
```

```python
# Scanpy preprocessing steps
import scanpy as sc
# Store raw counts in layers
adata.layers['counts'] = adata.X.copy()

# Normalize, scale and compute pca
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.scale(adata, max_value=10)
sc.tl.pca(adata)

# Return raw counts to X
dc.pp.swap_layer(
    adata=adata,
    key='counts',
    inplace=True
)
```

```python
adata
```

```python
# delete non useful information, and also we already now the sampleID (obs names), the run (72), and time (4)
del adata.obs['sampleID']
del adata.obs['run']
del adata.obs['time']
```

```python
# Correlate PCA explained variance with our observations
dc.tl.rankby_obsm(adata, key='X_pca')
```

```python
# As expected, our "treatment" should what is significant
sc.pl.pca_variance_ratio(adata)
dc.pl.obsm(
    adata=adata,
    return_fig=True,
    nvar=5,
    titles=['PC scores', 'Adjusted p-values'],
    figsize=(6, 3)
)
```

```python
adata.obs['experiment'] = adata.obs['experiment'].astype('category')
```

```python
sc.pl.pca(
    adata,
    color=['treatment', 'experiment'], 
    ncols=3,
    size=300,
    frameon=True,
)
```

```python
# genes removed
# MTND2P30", "RPL23AP95", "USP9YP10
```

```python
# Import DESeq2
from pydeseq2.dds import DeseqDataSet, DefaultInference
from pydeseq2.ds import DeseqStats

# Build DESeq2 object
inference = DefaultInference(n_cpus=10)
dds = DeseqDataSet(
    adata=adata,
    design_factors='treatment', # we could add here experiment (in case you didn't correct the data)
    refit_cooks=True,
    inference=inference,
)

# Compute LFCs
dds.deseq2()

# Extract contrast between conditions
stat_res = DeseqStats(
    dds,
    contrast=["treatment", 'mix', 'zero'],  # i want to compare mix vs zero
    inference=inference
)

# Compute Wald test
stat_res.summary()
```

```python
# Extract results, output from DESeq2.
results_df = stat_res.results_df
results_df
```

```python
dc.pl.volcano(
    data=results_df,
    x='log2FoldChange',
    y='padj',
    top=10,
)
```

```python
# we use t.values as it accounts for significance of the change as well as the Log2FC 
data = results_df[['stat']].T.rename(index={'stat': 'mix.vs.control'})
data
```

```python
# load collectri GRN
collectri = dc.op.collectri(organism='human')
collectri
```

```python
# Estimate TF activities, postive mean active (overrepresented) based on their target genes expression. While negative (underrepresented)
tf_acts, tf_padj = dc.mt.ulm(data=data, net=collectri)

# Filter by sign padj
msk = (tf_padj.T < 0.05).iloc[:, 0]
tf_acts = tf_acts.loc[:, msk]


tf_acts
```

```python
# Top 25 over and underrepresented TFs.
dc.pl.barplot(
    data=tf_acts,
    name='mix.vs.control',
    top=25,
    figsize=(10, 5)
)
```

```python
#HIF1A, CREB1, STAT3, CEBPB, and ATF2 
```

```python
dc.pl.network(
    net=collectri,
    data=data,
    score=tf_acts,
    sources=['CEBPG',  'CREB1', 'CEBPB'],
    targets=15,
    figsize=(10, 10),
    vcenter=True,
    by_abs=True,
    size_node=20
)
```

```python
for tf in tf_acts.columns:
    dc.pl.volcano(
        data=results_df,
        x='log2FoldChange',
        y='pvalue',
        net=collectri,
        name=tf,
        top=10,
    )
```
