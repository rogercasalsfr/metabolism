```python
import pandas as pd
from igraph import Graph

edges_df = tf_net[['source', 'target', 'weight']]


# Create graph from list of tuples (source, target, weight)
g = Graph.TupleList(edges_df.itertuples(index=False, name=None), 
                    directed=True, 
                    edge_attrs=["weight"])

print(g.summary())

# Step 3 (optional): Set weights
g.es['weight'] = edges_df['weight'].tolist()

# Now you have the graph `g`, you can analyze or plot it
print(g.summary())

# Example: Get degree of each TF
degrees = g.degree()
names = g.vs['name']
degree_dict = dict(zip(names, degrees))
print(degree_dict)
```

```python
import matplotlib.pyplot as plt

# Obtenir el layout com coordenades
layout = g.layout("fr")
coords = [layout[v.index] for v in g.vs]

# Crear el plot
plt.figure(figsize=(12, 8))

# Dibuixar arestes
for e in g.es:
    source, target = e.tuple
    x0, y0 = coords[source]
    x1, y1 = coords[target]
    plt.plot([x0, x1], [y0, y1], 'k-', linewidth=abs(e['weight']))

# Dibuixar nodes
x, y = zip(*coords)
plt.scatter(x, y, s=100, c="skyblue", edgecolors="k")

# Afegir etiquetes
for i, name in enumerate(g.vs['name']):
    plt.text(x[i], y[i], name, fontsize=8, ha='center', va='center')

plt.axis('off')
plt.title("TF-TF Network (manual layout)")
plt.show()
```
